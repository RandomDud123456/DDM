```python
import scipy.stats
import pyddm
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pyddm.plot
```

### Congruent Trials

```python
model_c = pyddm.gddm(drift="driftrate", noise="noise", bound="B", starting_position="x0", nondecision="ndt",
                   parameters={"driftrate": (-1, 1), 
                               "B": (.5, 2),
                               "x0": (-.5, .5),
                               "ndt": (0, .5),
                               "noise": (0,1)},
                               T_dur=3)
data_c = pd.read_csv("data_c.csv")
sample_c = pyddm.Sample.from_pandas_dataframe(data_c, rt_column_name="rt", choice_column_name="correct")
model_c.fit(sample_c)
```

```python
pyddm.plot.model_gui_jupyter(model_c)
```
